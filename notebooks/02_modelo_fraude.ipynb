```python
# Código para ejecutar dentro del notebook de desarrollo
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import joblib
import os

# 1. Cargar datos de entrenamiento
df_siniestros = pd.read_csv("../data/raw/siniestros.csv")

# 2. Selección estricta de características cuantitativas
X = df_siniestros[['monto_reclamado', 'dias_desde_inicio_poliza', 
                   'dias_desde_fin_poliza', 'dias_entre_ocurrencia_reporte', 
                   'historial_siniestros_asegurado']].fillna(0)

y = df_siniestros['etiqueta_fraude_simulada'].fillna(0)

# 3. Entrenar modelo corporativo
model = RandomForestClassifier(n_estimators=100, max_depth=8, random_state=42)
model.fit(X, y)

# 4. Serializar y guardar el artefacto en la ruta de producción
os.makedirs("../src/models/saved_models", exist_ok=True)
joblib.dump(model, "../src/models/saved_models/fraud_model.pkl")
print("¡Modelo entrenado externamente y exportado con éxito en src/models/saved_models/fraud_model.pkl!")
```
